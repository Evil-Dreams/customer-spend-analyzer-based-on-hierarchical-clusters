```python
import kagglehub
from kagglehub import KaggleDatasetAdapter
data = kagglehub.dataset_load(
  KaggleDatasetAdapter.PANDAS,
  "vibishananandhan/mall-customer-classification-using-k-means",
  "Mall_Customers.csv")
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['Genre'] = le.fit_transform(data['Genre'])
```

```python
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt
```

```python
X = data[['Annual_Income_(k$)', 'Spending_Score']].values
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5, linkage = 'ward')
y_hc = hc.fit_predict(X)
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(X[y_hc == 3, 0], X[y_hc == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(X[y_hc == 4, 0], X[y_hc == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')
plt.title('Agglomerative')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```

```python
from sklearn.cluster import KMeans
import numpy as np

def recursive_kmeans_divisive(data, n_clusters_target, random_state=42):
    clusters = {0: data}
    cluster_assignments = np.zeros(data.shape[0], dtype=int)

    current_cluster_id = 0
    next_new_cluster_id = 1
    while len(clusters) < n_clusters_target:
        cluster_to_split_id = max(clusters, key=lambda k: len(clusters[k]))
        cluster_to_split_data = clusters[cluster_to_split_id]

        if len(cluster_to_split_data) < 2:
            break
        kmeans = KMeans(n_clusters=2, random_state=random_state, n_init='auto')
        kmeans.fit(cluster_to_split_data)
        sub_assignments = kmeans.labels_
        original_indices_in_split_cluster = np.where(cluster_assignments == cluster_to_split_id)[0]
        new_cluster_1_data = cluster_to_split_data[sub_assignments == 0]
        new_cluster_2_data = cluster_to_split_data[sub_assignments == 1]
        cluster_assignments[original_indices_in_split_cluster[sub_assignments == 0]] = cluster_to_split_id
        cluster_assignments[original_indices_in_split_cluster[sub_assignments == 1]] = next_new_cluster_id
        clusters[cluster_to_split_id] = new_cluster_1_data
        clusters[next_new_cluster_id] = new_cluster_2_data
        reconstructed_clusters = {}
        for i in range(n_clusters_target):
            points_in_cluster = data[cluster_assignments == i]
            if len(points_in_cluster) > 0:
                reconstructed_clusters[i] = points_in_cluster
        clusters = reconstructed_clusters
        next_new_cluster_id += 1
    unique_labels = np.unique(cluster_assignments)
    mapping = {label: new_label for new_label, label in enumerate(unique_labels)}
    final_assignments = np.array([mapping[label] for label in cluster_assignments])
    
    return final_assignments

n_divisive_clusters = 5
y_divisive = recursive_kmeans_divisive(X, n_divisive_clusters)
plt.figure(figsize=(10, 7))
colors = ['red', 'blue', 'green', 'cyan', 'magenta', 'yellow', 'black', 'purple', 'orange', 'brown']
for i in range(n_divisive_clusters):
    plt.scatter(X[y_divisive == i, 0], X[y_divisive == i, 1], s = 100, c = colors[i % len(colors)], label = f'Cluster {i+1}')

plt.title('Divisive')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```
